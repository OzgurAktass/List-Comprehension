```python
import seaborn as sns
import pandas as pd
```

```python
df = pd.read_csv("/kaggle/input/car-crash/car_crashes.csv")
df.head()
```

```python
# ###############################################
# # TASK 1: Using the List Comprehension structure, capitalize the names of the numeric variables in the car_crashes data and append NUM to the beginning.
# ###############################################

# # Notes
# # # Non-numeric names should also grow.
# # Must be done with a single list comp structure.

df.columns

a = ["NUM_" + col.upper() if df[col].dtypes != "O" else col.upper() for col in df.columns]

a
```

```python
# ###############################################
# # TASK 2: Using the List Comprehension structure, write “FLAG” at the end of the names of the variables in the car_crashes data that do not contain “no” in their name.
# ###############################################
#
# # Notes
# # All variable names must be capitalized.
# # Must be done with a single list comp.

b = [col.upper() if "no" in col else col.upper() + "_FLAG" for col in df.columns]

b
```

```python
# ###############################################
# # Task 3: Using List Comprehension structure, select the names of the variables that are DIFFERENT from the variable names given below and create a new dataframe.
# ###############################################

og_list = ["abbrev", "no_previous"]

c = [col for col in df.columns if col not in og_list]

c

df[c].head()
```
